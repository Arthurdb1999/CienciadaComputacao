```python
import math
import seaborn as sns
import collections as cl
import numpy as np
df_iris = sns.load_dataset("iris")
iris_classe = 'species'
```

```python
df_iris15 = df_iris.sample(15, random_state=1)
```

```python
def entropiaAll(valores):
    counter = cl.Counter(valores)
    vp = [ (n/len(valores)) for n in counter.values() ] 
    somatorio = 0
    for p in vp :
        somatorio += 0 if (p == 0) else p * np.log2(p)    
    return somatorio * (-1)
```

```python
entropiaDF = entropiaAll(df_iris15)
entropiaDF
```

```python
df_iris15a = df_iris15[['sepal_length', iris_classe]]
df_iris15a.sort_values(by='sepal_length')
```

<h2> Split Points </h2>
<h4>1. Entre 4.7 e 5.1 = (4.7 + 5.1)/2 = 4.9</h4>
<h4>2. Entre 5.1 e 5.1 = (5.1 + 5.1)/2 = 5.1</h4>
<h4>3. Entre 5.4 e 5.6 = (5.4 + 5.6)/2 = 5.5</h4>
<h4>4. Entre 5.6 e 5.8 = (5.6 + 5.8)/2 = 5.7</h4>
<h4>5. Entre 5.8 e 6.0 = (5.8 + 6.0)/2 = 5.9</h4>
<h4>6. Entre 6.6 e 6.9 = (6.6 + 6.9)/2 = 6.75</h4>

```python
sp1=4.9
sp2=5.1
sp3=5.5
sp4=5.7
sp5=5.9
sp6=6.75
df_iris15a.loc[:,('sepal_length <= sp1')] = df_iris15a['sepal_length'].apply(lambda x : x <= sp1).values
df_iris15a.loc[:,('sepal_length <= sp2')] = df_iris15a['sepal_length'].apply(lambda x : x <= sp2).values
df_iris15a.loc[:,('sepal_length <= sp3')] = df_iris15a['sepal_length'].apply(lambda x : x <= sp3).values
df_iris15a.loc[:,('sepal_length <= sp4')] = df_iris15a['sepal_length'].apply(lambda x : x <= sp4).values
df_iris15a.loc[:,('sepal_length <= sp5')] = df_iris15a['sepal_length'].apply(lambda x : x <= sp5).values
df_iris15a.loc[:,('sepal_length <= sp6')] = df_iris15a['sepal_length'].apply(lambda x : x <= sp6).values
df_iris15a
```

```python
entropiaSP1 = 1/15 * - (1 * np.log2(1)) + 14/15 * - (4/14 * np.log2(4/14) + 6/14 * np.log2(6/14) + 4/14 * np.log2(4/14))  
print("Entropia SP1: ", entropiaSP1)
entropiaSP2 = 3/15 * - (1/3 * np.log2(1/3) + 2/3 * np.log2(2/3)) + 12/15 * - (3/12 * np.log2(3/12) + 5/12 * np.log2(5/12) + 4/12 * np.log2(4/12))
print("Entropia SP2: ", entropiaSP2)
entropiaSP3 = 5/15 * - (4/5 * np.log2(4/5) + 1/5 * np.log2(1/5)) + 10/15 * - (1/10 * np.log2(1/10) + 5/10 * np.log2(5/10) + 4/10 * np.log2(4/10))
print("Entropia SP3: ", entropiaSP3)
```

```python
print("Ganho SP1: ", entropiaDF - entropiaSP1);
print("Ganho SP2: ", entropiaDF - entropiaSP2);
print("Ganho SP3: ", entropiaDF - entropiaSP3);
print("Ganho SP4: ", entropiaDF - entropiaSP4);
print("Ganho SP5: ", entropiaDF - entropiaSP5);
print("Ganho SP6: ", entropiaDF - entropiaSP6);
```
