```python
def entropia (vp):
    somatorio = 0;
    for p in vp :
        somatorio += 0 if (p == 0) else p * np.log2(p)    
    return somatorio * (-1)

  
def entropia_valores(valores):
    import collections
    counter = collections.Counter(valores)
    vp = [ (n/len(valores)) for n in counter.values() ] 
    return entropia(vp)
```

```python
entropia_valores(df['PlayTennis'])
```

```python
def entropia_atributo(df,atributo,classe) :
      somatorio = 0    
      for valor, Xj in df.groupby(atributo):        
          p_Xj = len(Xj) / len(df)        
          E_Xj = entropia_valores(Xj[classe])
          somatorio += p_Xj * E_Xj        
      return somatorio
    
```

```python

for col in columns:
  
  print(entropia_atributo(df,col,'PlayTennis'))
  
```

# **Exercício - 1 - (c)**

```python
def ganho_de_informacao(df,atributo,classe):
  ent_classe = entropia_valores(df[classe])
  ent_atri= entropia_atributo(df,atributo,classe)
  return ent_classe - ent_atri
```

```python
for col in columns:
  print(ganho_de_informacao(df,col,'PlayTennis'))
```

# **Exercício -1 - (d)**

```python
def razao_IG(df,atributo,classe):
  IG = ganho_de_informacao(df,atributo,classe)
  I = entropia_valores(df[atributo])
  return IG / I
```

```python
for col in columns:
  print(razao_IG(df,col,'PlayTennis'))
```
